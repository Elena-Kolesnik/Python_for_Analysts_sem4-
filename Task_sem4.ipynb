```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv('kc_house_data.csv')
df.head()
```

***1.2 Изучите стоимости недвижимости***

```python
display(df['price'].describe())
print(f"Количество значений цены домов больше 2 миллионов: {(df['price'] > 2_000_000).sum()}")
```

Большинство объектов недвижимости стоят меньше двух миллионов рублей.


```python
plt.hist(df['price'], bins=150)
plt.xlim(0, 2_000_000)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.xticks(rotation=90)
plt.show()
```

***1.3 Изучите распределение квадратуры жилой***

```python
display(df['sqft_living'].describe())
print(f"Количество значений домов с жилой площадью более 5.000 кв.м: {(df['sqft_living'] > 5_000).sum()}")
```

```python
plt.hist(df['sqft_living'], bins=150)
plt.xlim(0, 5_000)
plt.title('Распределение жилой квадратуры')
plt.xlabel('Квадратные метры')
plt.ylabel('Частота')
plt.show()
```

Основная часть домов имеет квадратную жилую площадь в интервале от 1200 до 2700

***1.4 Изучите распределение года постройки***

```python
plt.hist(df['yr_built'], bins=50)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Частота')
plt.show()
```

***Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.***

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.scatter(df['price'], df['sqft_living'])
plt.title('Взаимосвязь между ценой и жилой площадью')
plt.xlabel('Цена в миллионах')
plt.ylabel('Жилая площадь')
plt.show()
```

scatterplot:

большинство домов находятся в пределах 2.5 миллионов и до 5.000 квадратных метров.
Имеется также интересный выброс: дом стоимостью около 2,5 млн рублей с жилой площадью 14000 кв м.

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg')
plt.title('Взаимосвязь между ценой и жилой площадью')
plt.xlabel('Цена в миллионах')
plt.ylim(0, 10000)
plt.ylabel('Жилая площадь')
plt.show()
```

jointplot:

чем больше жилая площадь, тем выше цена. 


```python
import numpy as np

corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt.figure(figsize=(20, 20))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```

heatmap:

признаки, имеющие взаимосвязь с ценой:

batrooms 0.5
sqft_above 0.6
sqft_living15 0.6
sqft_living 0.7
grade 0.7


```python
sns.barplot(x=df['grade'], y=df['price']); 
```

Чем выше оценка, тем больше стоит жильё.

```python
plt.figure(figsize=(8, 8))
d = df.groupby('grade').agg({"price": "count"})
d = d[d['price'] > 200]
plt.title('Продажи в зависимости от оценки жилой ед.')
plt.pie(d['price'],
        labels=d.index,
        autopct='%1.1f%%',
        wedgeprops={ 'linewidth' : 5, 'edgecolor' : 'white' }
        );
```

Самое продаваемое жилье 7 и 8 класса.
